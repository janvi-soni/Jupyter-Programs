```python
xbar=df["Engine Size(L)"].mean()
xbar
```

```python
ybar=df["CO2 Emissions(g/km)"].mean()
ybar
```

```python
df[["numeratorr"]]=(df["Engine Size(L)"]-xbar)*(df["CO2 Emissions(g/km)"]-ybar)
df
```

```python
numerator=df["numeratorr"].sum()
numerator
```

```python
df[["denominatorr"]]=(df["Engine Size(L)"]-xbar)*(df["Engine Size(L)"]-xbar)
df
```

```python
denominator=df["denominatorr"].sum()
denominator
```

```python
theta1=numerator//denominator
theta1
```

```python
theta0=ybar-(theta1*xbar)
theta0
```

```python
predicted_value=theta0+(theta1*df["Engine Size(L)"])
predicted_value
```

```python
# above prediction is manually(by using formula and all) but now we will do prediction by predefined methods of scikit learn
from sklearn.linear_model import LinearRegression   # provide a linear regression model for prediction
from sklearn.metrics import r2_score               # for finding accuracy
from sklearn.model_selection import train_test_split #split the dataset for training and testing
from sklearn.preprocessing import StandardScaler,PolynomialFeatures # standardscalar-for scaling the data after performing standardization and all
```

```python
x=df[["Engine Size(L)"]] #x contains the whole column  of engine size
y=df[["CO2 Emissions(g/km)"]] # y contain the whole column of c02 emission
slr=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=6200)# split the data for training and testing
```

```python

ss=StandardScaler() #object creation
x_train=ss.fit_transform(x_train)
```

```python
slr.fit(x_train,y_train) #train the model
x_test=ss.fit_transform(x_test) #fit find the mean and variance etc and transformmethod transform the testingdata of engine size according to mean 
ypred=slr.predict(x_test) #predict method takes the transformed data of feature(engine_size) and generate a response(co2_emission)
ypred
```

```python
accuracyoftrainmodel=slr.score(x_train,y_train)
accuracy=r2_score(ypred,y_test)
```
